# Density Matrix Simulation Notebook

Notebook version of `main.py`.


```python
from runner import run, loop_run
from datadump import dump_result_to_csv
from visualization import (
    plot_density_matrix_heatmap_plotly,
    plot_multi_scatter_from_csv_plotly,
    plot_scatter_from_csv_plotly,
)
import numpy as np
```

```python
# Parameters (equivalent to argparse defaults / examples in main.py)
bits = 3
p1 = 0.01
p2 = 0.1
pairs = [(0, 1), (1, 2)]
```

```python
# Run GHZ/W simulation and plot density-matrix heatmaps
results = run(bits=bits, p1=p1, p2=p2, pairs=pairs)
plot_density_matrix_heatmap_plotly(results, width=800, height=1200)
```

```python
# Parameter sweep and CSV dump
p2_sweep = np.arange(0.0, 0.62, 0.02, dtype=float)
loop_results = loop_run(
    bits=3,
    pairs=[(0, 1), (1, 2)],
    sweep_param="p2",
    sweep_values=p2_sweep,
    p1=0.0,
)
df = dump_result_to_csv(loop_results, "out/loop_metrics.csv")
df.head()
```

```python
# Multi-scatter subplot example: GHZ vs W-state
plot_multi_scatter_from_csv_plotly(
    csv_path="out/loop_metrics.csv",
    x_axes="sweep_value",
    data_sets=[
        ("w_concurrence_ab_pure", "ghz_concurrence_ab_pure"),
        ("w_entropy_pure", "ghz_entropy_pure"),
        ("w_monogamy_pure", "ghz_monogamy_pure"),
    ],
    x_label="p2",
    y_labels=["Concurrence", "von Neumann Entropy", "Monogamy"],
    titles=[
        "Concurrence Comparison",
        "von Neumann Entropy Comparison",
        "Monogamy Comparison",
    ],
    figure_title="W vs GHZ Metric Sweep",
    width=600,
    height=1000,
)
```
